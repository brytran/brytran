```python
import pandas
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tokenize import String
import csv

# project idea from kaggle, Titanic Machine Learning from Disaster
# https://www.kaggle.com/competitions/titanic/overview
# 60% accuracy rating, needs further improvement
```

```python
with open("train.csv", 'r') as inp, open("trainedit.csv", 'w') as out:
    writer = csv.writer(out)
    count = 0
    missing_data = 0
    index = [1,2,5,6,7,9]
    age = 0
    for line in inp:
        check = []
        status = "good"

        line = line.strip()
        temp = line.split(",")


        if count == 0:
            count += 1
            for num in index:
                check.append(temp[num])
            writer.writerow(check)
            continue

        temp[3] = temp[3] + temp[4]
        temp.pop(4)

        for num in index:
            value = temp[num]
            if value == "":
                status = "bad"
                continue
            check.append(float(temp[num]))
            
        try:  
            age += int(temp[5])
            count += 1
        except:
            missing_data += 1
        
        if len(check) != 6:
            status = "bad"

        if status == "bad":
            continue
        writer.writerow(check)

average_age = round(age / (count - missing_data), 2)
                
average_age


```

```python
with open("test.csv", "r") as inp, open("testedit.csv", "w") as out:
    writer = csv.writer(out)
    count = 0
    index = [1,4,5,6,8]
    for line in inp:
        check = []
        status = "good"

        line = line.strip()
        temp = line.split(",")

        if count == 0:
            count += 1
            for num in index:
                check.append(temp[num])
            writer.writerow(check)
            continue

        temp[2] = temp[2] + temp[3]
        temp.pop(2)

        for num in index:
            value = temp[num]
            try:
                check.append(float(temp[num]))
            except:
                check.append(average_age)


        writer.writerow(check)
        
test = pandas.read_csv("testedit.csv")
test
```

```python
fp = pandas.read_csv("trainedit.csv")
X = fp.drop(columns = "Survived")
Y = fp["Survived"]
X
```

```python
model = DecisionTreeClassifier()
model.fit(X,Y)
predictions = model.predict(test)
predictions
```

```python
with open("predictions.csv", "w") as pred:
    header = ['PassengerId', 'Survived']
    writer = csv.writer(pred)
    writer.writerow(header)
    ID = 892
    for num in predictions:
        pair = [ID, int(num)]
        writer.writerow(pair)
        ID +=1 
        
    
```
